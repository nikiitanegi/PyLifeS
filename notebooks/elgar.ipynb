```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('../datasets/elgar_appendix.csv')

# Create scatter plot
weight_column = 'weight (g)'
bmr_column = 'BMR'

plt.scatter(df[weight_column], df[bmr_column], alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('log of Weight (g)')
plt.ylabel('log of BMR')
plt.grid(True)

plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../datasets/elgar_appendix.csv')

weight_column = 'weight (g)'
bmr_column = 'BMR'

plt.plot(df[weight_column], df[bmr_column], marker='o')

plt.xlabel('weight (g)')
plt.ylabel('BMR')
plt.title(' Plot')
plt.grid(True)
plt.tight_layout()
plt.show() 
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('../datasets/elgar_appendix.csv') 

weight_column = 'weight (g)'
bmr_column = 'BMR'

log_weight = np.log(df[weight_column])
log_bmr = np.log(df[bmr_column])

plt.scatter(log_weight, log_bmr, alpha=0.5)
plt.title('Scatter Plot of Log(Weight) vs. Log(BMR)')
plt.xlabel('Log of Weight (grams)')
plt.ylabel('Log of BMR')
plt.grid(True)
plt.show() 
```
